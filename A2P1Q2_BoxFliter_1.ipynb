```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#
img_raw = cv.imread('test.jpg')
#
img_raw = cv.cvtColor(img_raw, cv.COLOR_BGR2RGB)
#
plt.imshow(img_raw)
```

```python
#
rows,cols = img_raw.shape[:2]
```

```python
#
rows,cols
```

```python
#,
m = n = 3
buff = []
```

```python
img_raw[0,0]
```

```python
#bf
A = []#
for i in range(0,rows-n+1):
    #buff
    for j in range(0,cols):
        if(i == 0):
            buff.append(img_raw[i:i+n,j])
        else:
            buff[j] -= img_raw[i-1,j]
            buff[j] += img_raw[i+n-1,j]
    sum = []
    #
    for j in range(0,cols-m+1):
        if(j == 0):
            sum.append([0,0,0])
            for k in range(0,m):
                sum[j] += buff[k]
        else:
            sum.append([0,0,0])
            sum[j] = sum[j-1] - buff[j-1] + buff[j+m-1]
        A.append(sum[j])
```
