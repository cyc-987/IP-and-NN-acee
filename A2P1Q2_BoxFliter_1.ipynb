```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#读取图像
img_raw = cv.imread('test.jpg')
#颜色转换
img_raw = cv.cvtColor(img_raw, cv.COLOR_BGR2RGB)
#输出图像
plt.imshow(img_raw)
```

```python
#获取宽度和高度信息
rows,cols = img_raw.shape[:2]
```

```python
#查看长宽值
rows,cols
```

```python
#设定矩形模版,开辟数组存储中间变量
m = n = 3
buff = []
```

```python
img_raw[0,0]
```

```python
#进行bf主操作并丢进数组里
A = []#这是存放结果的数组
for i in range(0,rows-n+1):
    #创建或更新buff节点
    for j in range(0,cols):
        if(i == 0):
            buff.append(img_raw[i:i+n,j])
        else:
            buff[j] -= img_raw[i-1,j]
            buff[j] += img_raw[i+n-1,j]
    sum = []
    #计算每个部分和
    for j in range(0,cols-m+1):
        if(j == 0):
            sum.append([0,0,0])
            for k in range(0,m):
                sum[j] += buff[k]
        else:
            sum.append([0,0,0])
            sum[j] = sum[j-1] - buff[j-1] + buff[j+m-1]
        A.append(sum[j])
```
